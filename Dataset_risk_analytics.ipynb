```python
data = df_no_nan.drop(['product', 'delq_history'], axis = 1)
data
```

```python
x = data.iloc[:, 0:-1]
y = data.iloc[:,-1]
x
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify = y)
```

```python
x_train
```

```python
y_train
```

```python
clf = svm.SVC()
clf.fit(x_train, y_train)
y_predicted = clf.predict(x_test)
```

```python
#getting the accuracy 
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_predicted)
```

```python
accuracy
```

```python
# using random forest 
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

random_forest = RandomForestClassifier(max_depth=8, random_state=0)
random_forest.fit(x_train, y_train)
y_predicted_RF = random_forest.predict(x_test)
accuracy_RF = accuracy_score(y_test, y_predicted_RF)
```
