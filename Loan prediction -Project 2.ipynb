```python
#picking columns values for encoding. another way to write could have been df=data

df = data.iloc[:,:].values
```

```python
#encoding products to include in model-- [:], where the left space is the row, the right space is the column. in pythin, we count column starting with 0, hence product column is 6 and not 7.
```

```python
from sklearn.preprocessing import LabelEncoder 
new_prod = LabelEncoder()
record = new_prod.fit_transform(df[:,6])
data['prod'] = record
data
```

```python
#using If statement to group remaining_loan_term values
def group_rl_term (x):
    if x > 0 and x <= 20:
         return 0
    elif x > 20 and x <= 40:
         return 1
    elif x > 20 and x <= 60:
         return 2
    elif x > 60 and x <= 80:
         return 3
    else: 
         return 4
  
        
group_rl_term(20)

```

```python
#using If statement to group original_loan_term values 

def group_ol_term (x):
    if x > 0 and x <= 20:
         return 0
    elif x > 20 and x <= 40:
         return 1
    elif x > 40 and x <= 60:
         return 2
    elif x > 60 and x <= 80:
         return 3
    else: 
         return 4
  
        
group_ol_term(200)
```

```python
# Inspecting columns after grouping 

data['r_term'] = data.remaining_loan_term.apply(group_rl_term)
data
```

```python
data['o_term'] = data.original_loan_term.apply(group_ol_term)
data
```

```python
# test for duplicates in dataset

data.duplicated().sum()
```

```python
# modifying the data to eliminate duplicates

data = data.drop_duplicates(subset=['loan_id'])
```

```python
# dealing with numeric null values to means and objects to 0

data = data.fillna({
    'monthly_income' : data['monthly_income'].mean(),
    'TOB_months': data['TOB_months'].mean(),
    'closing_principal_balance': data['closing_principal_balance'].mean(),
    'original_loan_amount': data['original_loan_amount'].mean(),
    'delq_history': '0,0,0,0,0,0',  
})

data.isnull().sum() # replacing null values w ith the mean(for int./float)
```

```python
#Using this loop to pick the P-values as a bases for excluding columns that are not significant 

from scipy import stats

corr_df = pd.DataFrame(columns=['r', 'P-value'])

for col in data:
    if pd.api.types.is_numeric_dtype(data[col]):
        r , p = stats.pearsonr(data.target, data[col])
        corr_df.loc[col] = [round(r, 4), round(p, 4)]
    
corr_df
```

```python
# preparing data for split

y = data.target.values
x = data.drop(['loan_id','delq_history','product','original_loan_term','r_term','remaining_loan_term','target'], axis='columns')
```

```python
#inspecting values
x
```

```python
# Split data to train set and test set... keep 20% (0.2) to test, then 80%(0.8) for train

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=42)
```

```python
# build the actual data model and fit to training data

from sklearn import linear_model as lr 
training_model = lr.LogisticRegression().fit(x_train, y_train)
```

```python
# test model accuracy

lr_score = training_model.score(x_train,y_train)
print(f"Model accuracy: {lr_score}")
print(f"intercept: {training_model.intercept_}")    
print(f"slope: {training_model.coef_}")

```

```python
# test model accuracy

lr_score = training_model.score(x_test,y_test)
print(f"Model accuracy: {lr_score}")
print(f"intercept: {training_model.intercept_}")    
print(f"slope: {training_model.coef_}")
```

```python
#confusion matix: to test if any data got predicted wrongly

from sklearn.metrics import confusion_matrix

y_prediction = training_model.predict(x_test)
y_actual = y_test
cm = confusion_matrix(y_actual, y_prediction)

# heatmap visulization of confusion matrix
f, ax = plt.subplots(figsize =(7, 7))
sns.heatmap(cm, cmap='Greens', annot = True, linewidth=1, linecolor="red", fmt =".0f", ax=ax)
plt.xlabel("prediction")
plt.ylabel("actual")
plt.show()
```

```python
#importing Libraries to support model check
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
```

```python
models =  ["Nearest Neighbor", "Random Forest", "Polymomial SVM", "Decision Tree","AdaBoost", "Naive bayes"]

classifiers = [
    KNeighborsClassifier(3),
    RandomForestClassifier(max_depth=5, n_estimators=100),
    SVC(kernel="linear", C=0.025),
    SVC(kernel="poly", degree=3, C=0.025),
    SVC(kernel="rbf",C=1, gamma=2),
    DecisionTreeClassifier(max_depth=5),
    AdaBoostClassifier(n_estimators=100),
    GaussianNB()]
```

```python
models
```

```python
scores = []
for models, clf in zip (models, classifiers):
    clf.fit(x_train, y_train)
    score = clf.score(x_test, y_test)
    scores.append(score)
```

```python
df = pd.DataFrame()
df['Test Algorithm'] = models
df['Model Score'] = scores
df
```
